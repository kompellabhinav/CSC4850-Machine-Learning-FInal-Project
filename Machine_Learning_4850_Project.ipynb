The dataset contains 3 targets, Dropout, Graduated, and Enrolled. We decided to change the 3 targets into 2, Dropout and No_Dropout, by grouping Graduated and Enrolled together into No_Dropout. We then converted the targets to 0 and 1 for easier processing. We also rounded the Curricular units 1st sem (grade) and Curricular units 2nd sem (grade) columns to 3 decimal places.

```python
from sklearn.preprocessing import LabelEncoder

# A copy of the dataset so the original is unchanged
df=pd.read_csv('dataset.csv', encoding='utf-8')

display(df)

# Changing the entries Graduate and Enrolled to N0_Droput
for i in range(0,len(df['Target'])):
  if (df['Target'][i] == "Graduate") | (df['Target'][i] == "Enrolled"):
    df['Target'][i] = "No_Dropout"

display(df)

# Converting the targets to 0 and 1
le = LabelEncoder()
label = le.fit_transform(df['Target'])

df.drop("Target", axis=1, inplace=True)
 
df["Target"] = label

display(df)

# Rounding up the ['Curricular units 1st sem (grade)'] and ['Curricular units 2nd sem (grade)'] columns
for i in range(0,len(df['Curricular units 1st sem (grade)'])):
  df['Curricular units 1st sem (grade)'][i] = round(df['Curricular units 1st sem (grade)'][i],3)
for i in range(0,len(df['Curricular units 2nd sem (grade)'])):
  df['Curricular units 2nd sem (grade)'][i] = round(df['Curricular units 2nd sem (grade)'][i],3)

display(df)
plt.hist(df['Target'])
plt.title('Target')
plt.show
```

#### Pipiline Setup

```python
from sklearn.model_selection import train_test_split

# Set seed
RANDOM = 1234

# Separate target from dataframe
X = df.drop("Target", axis=1)
y = df["Target"]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import Perceptron, LogisticRegression, LinearRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.model_selection import KFold
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier


import warnings
warnings.filterwarnings("ignore")

def pipe(X_train, X_test, y_train, y_test):

  dt_pipeline = Pipeline([('scaler', StandardScaler()), ('dt', DecisionTreeClassifier(random_state=RANDOM))])
  rf_pipeline = Pipeline([('scaler', StandardScaler()), ('rf', RandomForestClassifier(random_state=RANDOM))])
  p_pipeline = Pipeline([('scaler', StandardScaler()), ('p', Perceptron(random_state=RANDOM))])
  nb_pipeline = Pipeline([('scaler', StandardScaler()), ('nb', GaussianNB())])
  lr_pipeline = Pipeline([('scaler', StandardScaler()), ('lr', LogisticRegression(random_state=RANDOM))])
  linr_pipeline = Pipeline([('scaler', StandardScaler()), ('linr', LinearRegression())])
  linr_svc_pipeline = Pipeline([('scaler', StandardScaler()), ('linr', svm.SVC(kernel='linear', random_state=RANDOM))])
  rbf_svc_pipeline = Pipeline([('scaler', StandardScaler()), ('rbf', svm.SVC(kernel='rbf', random_state=RANDOM))])
  gb_pipeline = Pipeline([('scaler', StandardScaler()), ('gb', GradientBoostingClassifier(random_state=RANDOM))])
  mlp_pipeline = Pipeline([('scaler', StandardScaler()),('mlp', MLPClassifier(random_state=RANDOM))])
  knn_pipeline = Pipeline([('scaler', StandardScaler()),('knn', KNeighborsClassifier())])
  poly_svc_pipeline = Pipeline([('scaler', StandardScaler()), ('poly', svm.SVC(kernel='poly', random_state=RANDOM))])
  ada_pipeline = Pipeline([('scaler', StandardScaler()),('ada',   AdaBoostClassifier(random_state=RANDOM))])

  AdaBoostClassifier(random_state=RANDOM)


  dt_pipeline.fit(X_train, y_train)
  rf_pipeline.fit(X_train, y_train)
  p_pipeline.fit(X_train, y_train)
  nb_pipeline.fit(X_train, y_train)
  lr_pipeline.fit(X_train, y_train)
  linr_pipeline.fit(X_train, y_train)
  linr_svc_pipeline.fit(X_train, y_train)
  rbf_svc_pipeline.fit(X_train, y_train)
  gb_pipeline.fit(X_train, y_train) 
  mlp_pipeline.fit(X_train, y_train)
  knn_pipeline.fit(X_train, y_train)
  poly_svc_pipeline.fit(X_train,y_train)
  ada_pipeline.fit(X_train,y_train)

  all_models = {
      'Decision Tree' : dt_pipeline,
      'Random Forest' : rf_pipeline,
      'Perceptron' : p_pipeline,
      'Guassian NB' : nb_pipeline,
      'Logistic Regression' : lr_pipeline,
      'Linear Regression' : linr_pipeline,
      'Linear SVM' : linr_svc_pipeline,
      'RBF SVM' : rbf_svc_pipeline,
      'Gradient Boosting' : gb_pipeline,
      'Multi Layer Perceptron': mlp_pipeline,
      'K Nearest Neighbor': knn_pipeline,
      'Poly SVM': poly_svc_pipeline,
      'AdaBoost': ada_pipeline
  }

  return all_models

def KfoldValidiation(X, y, split):
  kf = KFold(n_splits=10, shuffle=True, random_state=42)
  metrics_table = {'Precision':[], 'Recall':[], 'F1score':[], 'Accuracy':[]}
  model_names = [
      'Decision Tree',
      'Random Forest',
      'Perceptron',
      'Guassian NB',
      'Logistic Regression',
      'Linear Regression',
      'Linear SVM',
      'RBF SVM',
      'Gradient Boosting',
      'Multi Layer Perceptron',
      'K Nearest Neighbor',
      'Poly SVM',
      'AdaBoost'
  ]

  model_metrics = {name:{'Precision':[], 'Recall':[], 'F1score':[], 'Accuracy':[]} for name in model_names}
  model_metrics['Linear Regression'] = {'RMSE':[],'r2':[]}

  # Running each split
  for train_idx, test_idx in kf.split(X):
    X_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
    X_test, y_test = X.iloc[test_idx], y.iloc[test_idx]

    all_models = pipe(X_train, X_test, y_train, y_test)    # Calling pipeline to fit each model
    
    for name,model in all_models.items():
      model_metric_fold = class_report(name, model, X_test, y_test)

      for key in model_metrics[name]:
        model_metrics[name][key].append(model_metric_fold[key])

  for name, model in model_metrics.items():
    df = pd.DataFrame.from_dict(model)
    print(f'{name} (split={split}):')
    display(df.round(3))
    print('\n')

def class_report(name, model, X_test, y_test):
  
  y_pred = model.predict(X_test)

  if name == 'Linear Regression':
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    r2 = r2_score(y_test,y_pred)
    metrics = {'RMSE':rmse,'r2':r2}

  else:
    report = classification_report(y_test, y_pred, output_dict=True)
    precision = report['weighted avg']['precision']
    recall = report['weighted avg']['recall']
    f1_score = report['weighted avg']['f1-score']
    accuracy = accuracy_score(y_test, y_pred)
    metrics = {'Precision':precision, 'Recall':recall, 'F1score':f1_score, 'Accuracy':accuracy}

  return metrics
```

```python
## Code to run all three splits on all models

for split in [.5,.3,.2]:
  print(f'Split: Test Size = {split} Train Size = {1-split}')
  X_train_split, X_test_split, y_train_split, y_test_split = train_test_split(X,y,random_state=RANDOM,test_size=split)
  KfoldValidiation(X_train_split, y_train_split, split)
  print('-'*50,'\n')
```

```python
#The best folds per split per model
# Handpicked using the f-score
model_best_folds = {
      'Decision Tree': [1, 2, 7],
      'Random Forest': [3, 2, 1],
      'Perceptron': [3, 1, 6],
      'Guassian NB': [3, 2, 6], 
      'Logistic Regression': [3, 2, 2],
      'Linear Regression': [3, 2, 6], 
      'Linear SVM': [0, 2, 6], 
      'RBF SVM': [8, 2, 1], 
      'Gradient Boosting': [3, 2, 6], 
      'Multi Layer Perceptron': [3, 2, 6],
      'K Nearest Neighbor': [3,2,3],
      'Poly SVM': [3,2,6],
      'AdaBoost': [4,2,6]
     }
```

Learning Curve and Confusion Matrix


```python
from sklearn.model_selection import LearningCurveDisplay, learning_curve
from sklearn.metrics import confusion_matrix

def get_best_fold(X, y, best_fold):
  kf = KFold(n_splits=10, shuffle=True, random_state=42)

  train_idx, test_idx = list(kf.split(X))[best_fold-1]
  X_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
  X_test, y_test = X.iloc[test_idx], y.iloc[test_idx]

  return X_train, X_test, y_train, y_test

train_sizes = [.1, .2, .3, .4, .5, .6, .7, .8, .9, 1]

def plot_learning_curve(X,y,name,model,fold,split,ts=train_sizes):
  train_sizes, train_scores, test_scores = learning_curve(model, X, y, train_sizes=ts, scoring='neg_mean_squared_error')

  train_scores = -train_scores
  test_scores = -test_scores
  train_mean = np.mean(train_scores, axis=1)
  test_mean = np.mean(test_scores, axis=1)

  fig, (ax1,ax2) = plt.subplots(nrows=2, figsize=(7,12))
  
  fig.suptitle(f'{name} (Split={split}, Fold={fold})')

  ax1.plot(train_sizes, train_mean, color='blue', label='training error')
  ax1.plot(train_sizes, test_mean, color='red', label='validation error')

  last_x = train_sizes[-1]
  last_y = test_mean[-1]
  ax1.annotate(f'{last_y:.2f}', xy=(last_x, last_y), xytext=(last_x, last_y),
    bbox=dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.2'))

  last_x = train_sizes[-1]
  last_y = train_mean[-1]
  ax1.annotate(f'{last_y:.2f}', xy=(last_x, last_y), xytext=(last_x, last_y),
    bbox=dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.2'))

  
  ax1.set_xlabel('Num of Training Samples')
  ax1.set_ylabel('Error')
  ax1.set_title('Learning Curve')

  for side in ax1.spines:
    ax1.spines[side].set_color('black')
    ax1.spines[side].set_linewidth(1)

  ax1.grid(False)
  ax1.set_facecolor('white')
  ax1.legend(loc=0)

  return ax2

def display_confusion_matrix(X_test,y_true,name,model,fold,split,axx):
    if name == 'Linear Regression':
      print('no confusion matrix for', name)
    else:
      y_pred = model.predict(X_test)
      cf_matrix = confusion_matrix(y_pred,y_true)

      labels = [[0,0],[0,0]]

      for i in range(2):
        for j in range(2):
            x = cf_matrix[i][j]
            x = f'{x} ({x/sum(cf_matrix.flatten()): .1%})'
            labels[i][j] = x

      sns.heatmap(cf_matrix, annot=labels, cmap='Blues', fmt='', ax=axx)
      axx.set_title('Confusion Matrix')
      axx.set_xlabel('Predicted')
      axx.set_ylabel('True')
      plt.show()
      print('\n')
```

Plot learning curves and confusion matrices for the best fold for each split in all models

```python
for i,split in enumerate([.5, .3, .2]):
  X_train_split, X_test_split, y_train_split, y_test_split = train_test_split(X,y,random_state=RANDOM,test_size=split)

  for name, folds in model_best_folds.items():
    best_model_fold = folds[i]
    X_train_best, X_test_best, y_train_best, y_test_best = get_best_fold(X_train_split, y_train_split,best_model_fold) 
    
    all_models = pipe(X_train_best, X_test_best, y_train_best, y_test_best)
    model = all_models[name]

    ax2 = plot_learning_curve(X_train_best,y_train_best,name,model,best_model_fold,split)
    display_confusion_matrix(X_test_best,y_test_best,name,model,best_model_fold,split,ax2)
 
```

Method for determining the difference between training and validation error.

```python
def get_learning_curve_gap(X,y,model,ts=train_sizes):
  train_sizes, train_scores, test_scores = learning_curve(model, X, y, train_sizes=ts, scoring='neg_mean_squared_error')

  train_scores = -train_scores
  test_scores = -test_scores
  train_mean = np.mean(train_scores, axis=1)
  test_mean = np.mean(test_scores, axis=1)

  error_diff = test_mean - train_mean

  return error_diff, train_sizes
```

```python
# Plotting how well log reg vs naive bayes generalizes 
split = .5
best_model_fold = 3

nb_pipeline = Pipeline([('scaler', StandardScaler()), ('nb', GaussianNB())])
lr_pipeline = Pipeline([('scaler', StandardScaler()), ('lr', LogisticRegression(random_state=RANDOM))])

X_train_best, X_test_best, y_train_best, y_test_best = get_best_fold(X_train_split, y_train_split,best_model_fold) 

err1, ts1 = get_learning_curve_gap(X_train_best,y_train_best,nb_pipeline)
err2, ts2 = get_learning_curve_gap(X_train_best,y_train_best,lr_pipeline)

fig, ax = plt.subplots(figsize=(7,7))

plt.plot(ts1, err1, color='blue', label='Naive Bayes')
plt.plot(ts2, err2, color='red', label='Log Reg')

plt.xlabel('Num of Training Samples')
plt.ylabel('Generalization Error')
plt.title('Naive Bayes vs Log Reg (50-50 Split, Fold 3)')

plt.grid(False)
plt.legend(loc=0)
plt.show()
```
